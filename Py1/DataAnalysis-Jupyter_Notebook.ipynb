## Random Numbers and Creating Random Data

We usually work with read world data but we might need to make a dataset with random values.

```python
#importing libraries
import pandas as pd
from numpy import random
from numpy.random import randint

names = ['Bob', 'Jessica', 'Mary', 'John', 'Mel'] #List of names from which we will randomly select from
```

Seeding Random Generator

```python
random.seed(500)
```

###### Note: If we use the same seed, we will always get the same set of "random" numbers. 

```python
randnames = []
for i in range(1000): #Loops until i is equal to n, i.e 1,2,3...n
    name = names[randint(low=0, high=len(names))] #Generates a random number between zero and the length of the list names
    randnames.append(name) #Selects the random name selected and adds it to a list
    print(name) #Prints the name selected (for debugging)
```

To create a dataframe we now select 1000 random numbers

```python
births = []
for i in range(1000):
    births.append(randint(low = 0, high = 1000))
```

Finally zipping the two lists together creates a dataframe

```python
BabyDataSet2 = list(zip(randnames,births))
df = pd.DataFrame(data = BabyDataSet2,
                  columns=['Names','Births'])
df
```

### Calculating and Removing Outliers

Finding outliers allows you to remove the values that are so high or so low that they skew the overall view of the data.

1. Standard Deviations: If the data is normally distributed 95% of the data is within 1.96 standard deviations of the mean. So we can drop the values either above or below that range.
2. Interquartile Range (IQR): The IQR is the difference between 25 percentile and the 75 percentile. Any values that are either lower than Q1 - 1.5 x IQR or greater than Q3 + 1.5 x IQR are treated as outliers and removed.

```python
# Standard Deviation
import pandas as pd
Location = "D:\\Data Analysis\\Py1\\Datasets\\Walmart.csv"
df = pd.read_csv(Location)

meantemp = df['Temperature'].mean()
stdtemp = df['Temperature'].std()

toprange = meantemp + stdtemp * 1.96 #Calculate the upper range equal to 1.96 times the standard deviation plus the mean
botrange = meantemp - stdtemp * 1.96 #Calculate the standard deviation subtracted from the mean

copydf = df
copydf = copydf.drop(copydf[copydf['Temperature'] > toprange].index) #Drop the rows where the temperature is higher than the top range
copydf = copydf.drop(copydf[copydf['Temperature'] < botrange].index) #Drop the rows where the temperature is lower than the bottom range
copydf
```

```python
# Interquartile Range
import pandas as pd
Location = "D:\\Data Analysis\\Py1\\Datasets\\Walmart.csv"

df = pd.read_csv(Location)
q1 = df['Temperature'].quantile(.25)
q3 = df['Temperature'].quantile(.75)
iqr = q3 - q1
toprange = q3 + iqr * 1.5 #Calculate upper boundary = the third quartile + 1.5*IQR
botrange = q1 - iqr * 1.5 #Calculate lower boundary = the first quartile - 1.5*IQR

copydf = df
copydf = copydf.drop(copydf[copydf['Temperature'] > toprange].index) # Drop the rows where the temperature is higher than the toprange
copydf = copydf.drop(copydf[copydf['Temperature'] < botrange].index) # Drop the rows where the temperature is lower than the botrange
copydf
```

### Missing Data in Dataframes

When no information is provided for one or more elements or for the entire unit, this is referred to as missing data.

```python
# Creating Dataframes with missing data
import pandas as pd
df = pd.read_csv("D:\\Data Analysis\\Py1\\Datasets\\listings.csv")
df.head
```

```python
# Drop all rows with missing (NaN) data
df_no_missing = df.dropna()
df_no_missing
```

```python
# Add columns with empty values
import numpy as np
df['newcol'] = np.nan
df.head()
```

```python
#Replace empty cells with zero
df.fillna(0)
```

```python
# Replace empty cells with Average of Column
df["newcol"].fillna(df["newcol"].mean(), inplace = True)
df
```

```python
# Drop completely empty columns
df.dropna(axis=1, how='all')
```

```python
# To select rows without missing data
print(df["minimum_nights"].fillna(df.groupby("minimum_nights")["minimum_nights"].transform("mean"),inplace = True))
```

### Inappropriate Values

Sometimes we need to worry if the given data is accurate. Pandas library has tools to help and get rid of impossible or inappropriate values

```python
#Creating Dataset
import pandas as pd
names = ['Bob', 'Jessica', 'Mary', 'John', 'Mel']
grades = [76, -2, 77, 78, 101]

GradeList = zip(names,grades)
df = pd.DataFrame(data = GradeList,
                  columns=['Names', 'Grades'])
df
```

```python
#Filtering out Impossible grades by eliminating all grades where the grades are too high
df.loc[df['Grades'] <= 100]
```

```python
#Or change Impossible Grades
df.loc[(df['Grades'] >= 100, 'Grades')] = 100
df
```

```python
#Now first changing all sub zero grades to zero and then removing failures
df.loc[(df['Grades'] <= 0, 'Grades')] = 0
print(df)
df.loc[df['Grades']>= 1]
```
